```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
```

```python
list1=[]
def per(a):
    list1.append(round(a,2))
    return(format(a,'.1f'))
```

```python
plt.axis("equal")
plt.pie(sold,labels=company,autopct=per)
plt.show()
```
