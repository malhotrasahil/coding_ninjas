```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

x1 =[0,1,2,3,4,5,6,7,9,10]
y1=[1,2,5,10,17,26,37,50,82,101]
```

```python
x=np.arange(0,40,2)
y=x
```

```python
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("title")

plt.plot(x,y,color='blue',label="xy")
plt.plot(x1,y1,color='green',label="x1y1")
plt.legend()
plt.show()
```
